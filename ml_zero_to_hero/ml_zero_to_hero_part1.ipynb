ビデオ中にあった Y = 2X - 1 を学習させるコードの写経。

ビデオ撮影時は、 TensorFlow 1.x だったので、 kerasも独立している。 TensorFlow 2.x のコードは後述。

なお、ビデオ中にあった リンクは [Learn TensorFlow 1: The "Hello World" of machine learning](https://codelabs.developers.google.com/codelabs/tensorflow-lab1-helloworld/#0) でここには、公式が少し変わったが、 TensorFlow 2.x版のコードと説明がある。

## TensorFlow 1.x


```
% tensorflow_version 1.x
import keras
import numpy as np

model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)
print(model.predict([10.0]))
```

## TensorFlow 2.x

```
% tensorflow_version 2.x

import tensorflow as tf
import numpy as np
from tensorflow import keras

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)
print(model.predict([10.0]))
```
